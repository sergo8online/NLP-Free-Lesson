***Now lets write simple NN with RNN block***

```python
class TextRnn(nn.Module):
  def __init__(self, in_features, out_features):
    super().__init__()
    self.in_features = in_features
    self.out_features = out_features
    self.hidden_size = 64

    self.rnn = nn.RNN(input_size=self.in_features, hidden_size=self.hidden_size, batch_first=True)
    self.linear = nn.Linear(in_features = self.hidden_size, out_features = self.out_features)

  def forward(self, x):
    x, h = self.rnn(x)
    y = self.linear(h)
    return y
```

```python
import re
class CharsDataset(data.Dataset):
  def __init__(self, path, prev_chars=3): #path is our path for our text document, this text is pur dataset, prev_chars=3 means x=[a, b, c] and y = [d], so len(x) is prev_chars
    self.prev_chars = prev_chars

    with open(path, 'r') as f:
      self.text = f.read()
      self.text = self.text.replace("\ufeff", '')

    self.text = self.text.lower()
    self.alphabet = set(self.text) #unique characters from our text
    self.idx_to_char = dict(enumerate(sorted(self.alphabet)))   # something like that -> {0: 'a', 1: 'b', 2: 'c', 3: 'd', 4: 'e'}
    self.char_to_idx = {v:i for i, v in self.idx_to_char.items()} #something like that -> {'a': 0, 'b': 1, 'c': 2, 'd': 3, 'e': 4}
    self.num_char = len(self.alphabet) #how much characters do we have in our word corpus
    self.onehots = torch.eye(self.num_char) #our one hot encodings for each character is vector with lenght = self.num_chars, for first char its ve ctor [1, 0, 0....], for second [0, 1, 0...] and for last [0, 0, .....1] it means we can create them by using just I matrix, for this we used torch.eye

  def __getitem__(self, item): #this magic method must return as one sample from our training dataset
    _data = torch.vstack([self.onehots[self.char_to_idx[self.text[x]]] for x in range(item, item+self.prev_chars)])
    ch = self.text[item+self.prev_chars] #thats our target character
    label = self.char_to_idx[ch]   #and that its number representation
    return _data, label

  def __len__(self):
    return len(self.text) - 1 - self.prev_chars
```

```python
class TextRnnV1(nn.Module):
  def __init__(self, in_features, out_features):  #first is len of our embeddings, second in size of our vocab
    super().__init__()
    self.in_features = in_features
    self.out_features = out_features
    self.hidden_layers = 256

    self.rnn = nn.RNN(input_size=self.in_features, hidden_size=self.hidden_layers, batch_first=True)
    self.linear = nn.Linear(in_features=self.hidden_layers, out_features=self.out_features)

  def forward(self, x):
    x, h = self.rnn(x)
    y = self.linear(h)
    return y
```

```python
class CustomDataset(data.Dataset):
  def __init__(self, path, emb, tokenizer, prev_tokens=3):
    self.prev_tokens = prev_tokens
    self.emb = emb
    self.tokenizer = tokenizer

    with open(path, 'r') as f:
      self.text = f.read()
      self.text = self.text.replace("\ufeff", '')
      self.text = self.text.lower()                 #this is little preporccesing

    #lets do tokenization
    self.tokenized_text = self.tokenizer(self.text)  # -> return as a list of tokens
    self.vocab = set(self.tokenized_text)   # word corpus

    self.idx_to_token = dict(enumerate(sorted(self.vocab)))
    self.token_to_idx = {v:i for i, v in self.idx_to_token.items()}
    self.vocab_size = len(self.vocab)

  def __getitem__(self, item):
    _data = torch.vstack([torch.tensor(self.emb[self.tokenized_text[x]])  if self.tokenized_text[x] in self.emb else torch.zeros(50) for x in range(item, item+self.prev_tokens)])
    target = self.tokenized_text[item+self.prev_tokens]
    label = self.token_to_idx[target]
    return _data, label
  def __len__(self):
    return len(self.tokenized_text) - 1 - self.prev_tokens
```

```python
#if we wanna make our nn.RNN bidirectional, we should set parameter bidirectional=True
class BidirectionalRnn(nn.Module):
  def __init__(self, in_features, out_features):
    super().__init__()
    self.in_features = in_features
    self.out_features= out_features
    self.hidden_size = 16

    self.rnn = nn.RNN(self.in_features, self.hidden_size, batch_first=True, bidirectional=True)
    self.linear = nn.Linear(self.hidden_size*2, self.out_features)  #here we write hidden*2 becouse after rnn block, two h will be concatinated, it means the shape will be doubled

  def forward(self, x):
    x, h = self.rnn(x)
    hh = torch.cat((h[0, :, :], h[1, :, :]), dim=1)
    y = self.linear(hh)
    return y
```

```python
#we can use class LSTM from torch
#lets see how it works
import torch
import torch.nn as nn

lstm = nn.LSTM(10, 16, batch_first=True)  #10->size of each token's vector, 16-hidden_size
x = torch.randn(1, 5, 10)  #so shape of our x will be (batch_size, seq_lenght, d_size) d_size is the same size of each token's vector

y, (h, c) = lstm(x)

print(f"output y shape: {y.size()}") #idx=0 is batch
print(f"final hidden state's shape: {h.size()}") #idx=1 is batch
print(f"final context state's shape: {c.size()}") #idx = 1 is batch
print()

#what if we will use bidirectional lstm, how our shapes will change?
bi_lstm = nn.LSTM(10, 16, batch_first=True, bidirectional=True)
y_, (h_, c_) = bi_lstm(x)

print(f"output y shape: {y_.size()}")
print(f"final hidden state's shape: {h_.size()}")
print(f"final context state's shape: {c_.size()}")
```

in "putting all together" section you can change name nn.RNN to nn.LSTM and see what happens

> NOTE: dont forget that outputs from rnn and lstm are different



```python
class TextLSTM(nn.Module):
  def __init__(self, in_features, out_features):
    super().__init__()
    self.in_features = in_features
    self.out_features = out_features
    self.hidden_size = 64

    self.lstm = nn.LSTM(self.in_features, self.hidden_size)
    self.linear = nn.Linear(self.hidden_size, self.out_features)

  def forward(self, x):
    x, (h, c) = self.lstm(x)
    y = self.linear(h)
    return y
```

```python
class CustomDataset(data.Dataset):
  def __init__(self, texts, labels, emb, tokenizer, prev_tokens=None):
    super().__init__()
    self.labels = labels
    self.texts = [text.lower() for text in texts]   #every sentence in text must be lowercase
    self.emb = emb
    self.tokenizer = tokenizer

    #now we need to tokenize each sentence
    self.tokenized_texts = [self.tokenizer(sen) for sen in self.texts] #now self.tokenized_texts is a list full of another lists with tokens

    if prev_tokens == None:
      self.prev_tokens = max([len(l) for l in self.tokenized_texts])
    else:
      self.prev_tokens = prev_tokens

    # example_word = next(iter(self.emb))
    # self.emb_size = len(self.emb[example_word])  #size of each embedding
    self.emb_size = 50

  def __getitem__(self, item):
    #my getitem magic method must return as data and labels
    #labels will be the same
    #but now he must convert each list of tokens into list of embeddings, so each data point will be matrix (prev_tokens, embedding_size)
    tokens = self.tokenized_texts[item]
    #truncte
    tokens = tokens[:self.prev_tokens]

    embeddings = []
    for token in tokens:
      if token in self.emb:
        embedding = self.emb[token]
      else:
        embedding = np.zeros(self.emb_size)
      embeddings.append(torch.tensor(embedding, dtype=torch.float))

    while len(embeddings) < self.prev_tokens:
      embeddings.append(torch.zeros(self.emb_size))
    X = torch.stack(embeddings)
    y = torch.tensor(self.labels[item], dtype=torch.float)

    return X, y

  def __len__(self):
    return len(self.tokenized_texts)
```

```python
class TextClassificationRnn(nn.Module):
  def __init__(self, in_features, hidden_size):
    super().__init__()
    self.in_features = in_features
    self.hidden_size = hidden_size

    self.rnn = nn.RNN(self.in_features, self.hidden_size, batch_first=True)
    self.linear = nn.Linear(self.hidden_size, 1)

  def forward(self, x):
    y, h = self.rnn(x)
    pred = self.linear(h.squeeze(0))
    return pred

Rnn = TextClassificationRnn(50, 32)
```

```python
!pip install torchmetrics
from torchmetrics import Accuracy

acc_fn = Accuracy(task='binary')
loss_fn = nn.BCEWithLogitsLoss()
optimizer = torch.optim.Adam(params=Rnn.parameters())
```

```python
epochs = 20
for epoch in range(epochs):
  avg_loss, avg_acc = 0, 0
  Rnn.train()
  for X, y in train_dataloader:
    logits = Rnn(X).squeeze()
    loss = loss_fn(logits, y)
    acc = acc_fn(torch.round(torch.sigmoid(logits)), y)

    avg_loss += loss.item()
    avg_acc += acc.item()

    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
  avg_loss /= len(train_dataloader)
  avg_acc /= len(train_dataloader)
  print(f"Epoch: {epoch} | Loss: {avg_loss} | Acc: {avg_acc}")
```
